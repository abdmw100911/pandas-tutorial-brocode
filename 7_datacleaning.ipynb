```python
#data cleaning: the process of fixing or removing:
#                incomplete, incorrect or irrelevant data 
#                75% of work done through pandas is data cleaning 
import pandas as pd
```

```python
df=pd.read_csv("pokemon.csv")
print(df.to_string())
```

```python
#1 dropping irrelevant columns 
#suppose the column legendary is not relevant, so drop it

df=df.drop(columns=["Legendary"]) 
#this creates a new dataframe with the legendary column droppped
#this method uses labels to target columns and creates a new dataframe instead of modifiying the orginal one

print(df)
```

```python
#using index to target columns and modifying the original data frame
df.drop(df.columns[0], axis=1, inplace=True)
print(df)

# here we targeted the "#" column which was not possible through previous method
```

```python
#2. handling missing data 
#missing data can be handled by two methods
#a. dropping the entire row with the missing value
#b. replacing it with a different value 
print(df.to_string())
```

```python
#a. dropping the entire row with the missing value
#"Type 2 has some missing values for some pokemons, so we eliminate the pokemons who donot have a type 2"
df1=df.copy()
df.dropna(subset=["Type 2"],inplace=True)
#dropna=drop something with NaN value
#this will drop any row having missing value under the column type 2
#eg. charmendor, Charmeleon 

print(df.to_string())
```

```python
#b. replacing it with a different value
df1.fillna({"Type 2":"None"},inplace=True)
print(df1.to_string())
```

```python
#3. fixing inconsitent values
#suppose i want to change Grass to GRASS, Bug to BUG, Water to WATER

df1["Type 1"] = df1["Type 1"].replace({"Grass":"GRASS","Bug":"BUG","Water":"WATER"})
print(df1.to_string())
```

```python
#4. standardize text 
#suppose i want all the names "all of it " to be uppercase only 

df1["Name"]=df1["Name"].str.upper()
print(df1.to_string())
```

```python
#5. change /fix data types
df2=pd.read_csv("pokemon.csv")
#suppose i want to change legendary data type from "true" and "false" to "0" and "1"

df2["Legendary"]=df2["Legendary"].astype(int)

df2["Attack"]=df2["Attack"].astype(float)

print(df2.to_string())
```

```python
#6.removing duplicate values
#bulbasor is repeated so remove it

df2.drop_duplicates(inplace=True)
print(df2.to_string())
```

```python
#applying all the cleaning techniques to a single new data frame
df3=pd.read_csv("pokemon.csv")
print(df3.to_string())
#this is the uncleaned data, lets clean it
#1. "#" column is irrelevannt so remove it 
#2. drop rows who donot have a type 2 
#3. replace missing value of Total of some pokemons
#4. all the names of the pokemon must be uppercase
#5. Fire and water must be uppercase
#6. change legendary data type to be numeric
#7. remove duplicate values
```

```python
#1. "#" column is irrelevannt so remove it 
df3.drop(df3.columns[0],axis=1,inplace=True)
print(df3.to_string())
```

```python
#2. drop rows who donot have a type 2 
df3.dropna(subset=["Type 2"],inplace=True)
print(df3.to_string())
```

```python
#3. replace missing value of Total of some pokemons

df3.fillna({"Total":df3["Total"].mean()},inplace=True)
df3["Total"]=df3["Total"].astype(int)
print(df3.to_string())
```

```python
#4. all the names of the pokemon must be uppercase

df3["Name"]=df3["Name"].str.upper()
print(df3.to_string())
```

```python
#5. Fire and water must be uppercase

df3["Type 1"]=df3["Type 1"].replace({"Fire":"FIRE","Water":"WATER"})
print(df3.to_string())
```

```python
#6. change legendary data type to be numeric

df3["Legendary"]=df3["Legendary"].astype(int)
print(df3.to_string())
```

```python
#7. remove duplicates

df3.drop_duplicates(inplace=True)
print(df3.to_string())

#finally we have our clean and tidy data
```
