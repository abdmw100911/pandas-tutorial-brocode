```python
import pandas as pd
df=pd.read_csv("pokemon.csv")
print(df.iloc[:,:])
```

```python
#aggregation: summarizes the data ,
            # often used with groupby() function
```

```python
#aggregation functions for the whole dataframe

#1. mean function
#numeric_only=True, allows aggregate functions to be applied to columns with numeric value
print(df.mean(numeric_only=True)) 
```

```python
#2. sum function
print(df.sum(numeric_only=True))
```

```python
#3. min function
print(df.min(numeric_only=True))
```

```python
#4. max function
print(df.max(numeric_only=True))
```

```python
#5. count function:gives the count of elements in the column, doesnt consider NaN in the count
print(df.count())
#count can be applied to non numeric values as well, so no need to apply it for 
#numeric values only

#not all pokemons have a type 2, so type 2 is not 800
```

```python
#aggregate functions for a single column
#1. mean 
print(df["Speed"].mean())
```

```python
#2. sum
print(df["Sp. Atk"].sum())
```

```python
#3. min
print(df["Sp. Def"].min())
print(df["Defense"].min())
```

```python
#4. max
print(df["Sp. Atk"].max())
print(df["Attack"].max())
```

```python
#5. count
print(df["Type 2"].count())
```

```python
#using aggregate functions with group by 
#for eg here we can create different groups of the pokemon based on their type
group=df.groupby("Type 1") 

print(group["Attack"].mean().reset_index(name="mean_attack"))
```

```python
print(group["Attack"].max().reset_index(name="max_attack"))
#this will tell the maximum attack that each type has 
#eg. in bug type pokemon, the maximum attack is 185
#for Fire type, the maximum attack is 160 
```

```python
print(group["Attack"].min().reset_index(name="minimum_attack"))
#here it shows minimum attack of each type
```

```python
print(group["Attack"].count())
```

```python
#printing the name of the pokemon which has the highest attack in the group
grouped = df.groupby("Type 1")

max_attack_for_each_type = grouped["Attack"].idxmax()
# This returns the ROW INDEX of the Pok√©mon with the highest Attack inside each Type 1 group

display = df.loc[max_attack_for_each_type, ["Name", "Type 1", "Attack"]]
# Using those row indexes, we fetch the actual rows and only show specific columns

print(display)
```

```python
#finding the name of the pokemon with the minimum attack from each type 1
group_type=df.groupby("Type 1")

idx_of_min_attack=group_type["Attack"].idxmin()

display=df.loc[idx_of_min_attack,["Name", "Type 1", "Attack"]]

print(display)
```
