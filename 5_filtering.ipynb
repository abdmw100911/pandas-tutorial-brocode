```python
import pandas as pd 
```

```python
#filtering: keeping the rows that match a specific condition
df=pd.read_csv("pokemon.csv")
print(df.loc[0:10])
```

```python
#1. keeping all the pokemons whose speed is >= 100
fast_pokemon=df[df["Speed"]>=100]
print(fast_pokemon)
```

```python
#2. keeping all the legendary pokemons
legendary=df[df["Legendary"]==True]
print(legendary.loc[:,["Name","Legendary"]])
```

```python
#3.keeping all the powerfull pokemons(they have high  attack and defense)
powerful_pokemons=df[(df["Attack"]>=150) & (df["Defense"]>=150)]
print(powerful_pokemons.loc[:,["Name","Attack","Defense"]])
```

```python
#keeping all the water pokemons
water_pokemons=df[(df["Type 1"]=="Water") | (df["Type 2"]=="Water")]
print(water_pokemons.loc[:,["Name","Type 1","Type 2"]].to_string())
```
