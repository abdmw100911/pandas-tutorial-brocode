```python
import pandas as pd
df=pd.read_csv("pokemon.csv")
print(df)
```

```python
#SELECTION BY COLUMN
# print(df["Name"])
# print(df["Type 1"])

print(df[["Name","Type 1","Type 2","Legendary"]])
```

```python
#selection by rows
#selection using the loc property
print(df.loc[0])
print("\n")
print(df.loc[10])
```

```python
#since it is difficult to remember the pokemon through its index
#we can set the name column to be as index and then access information through name

df=pd.read_csv("pokemon.csv", index_col="Name")
print(df.loc["Pikachu"])
```

```python
#we can also select the specific column range we need to be printed 
#like i want column from "type 1 " to "defense " to be printed for pikachu

print(df.loc["Pikachu","Type 1":"Defense"])
```

```python
#we can also select specfic columns to be printed of the required pokemon 
print(df.loc["Pikachu",["Type 1","Type 2"]])
```

```python
#we can print information for specified rows and columns
print(df.loc[["Pikachu","Charizard"],["Type 1","Type 2"]])
```

```python
#we can print information for range of rows and columns at the same time 
print(df.loc["Bulbasaur":"Charmander","Type 1":"Defense"])
```

```python
#the same functions can be done through the "iloc" property as well
print(df.iloc[0])
```

```python
print(df.iloc[0,3])
#will print first row and 3rd column data
```

```python
print(df.iloc[[0,1,2,3],[1,2,3,4]])
```

```python
print(df.iloc[0:4,1:5])
```

```python
#also if you want every second row in the range , you can do that as well
print(df.iloc[0:20:2,0:5])
```
