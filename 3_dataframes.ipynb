```python
#dataframe: is a 2d tabular dataset with rows and columns, similar to an excel sheet
import pandas as pd

#creating a data frame through a dictionary 
emp_data={
    "name":["abd","hus","has","zain"],
    "age":[20,13,12,18],
    "salary":[1000,2000,4000,2000]
}
df=pd.DataFrame(emp_data)
print(df)
```

```python
#changing the index of the values in the dataframe
df=pd.DataFrame(emp_data,index=["emp1","emp2","emp3","emp4"])
print(df)
```

```python
#accessing each row(all information about a single employee)
#1. using loc(location by label)

print(df.loc["emp1"])

#2. using iloc(location by index)
print("\n")
print(df.iloc[3])
```

```python
#adding a new column or columns to existing data frame
print(df)
print("\n")
df["job"]=["engineer","marketing","relation","finance"]
print(df)
```

```python
#adding multiple new columns to the existing dataframe 
df[["exp", "loc"]]=pd.DataFrame({
    "exp" : [1,2,3,4],
    "loc":["mum","pune","bangalore","andhra"]
},
index=df.index)
print(df)
```

```python
#adding a new row to the existing dataframe
#1. create a new dataframe having each column and its value
new_row=pd.DataFrame([{
    "name":"zoh",
    "age":40,
    "salary":1000,
    "job":"ceo",
    "exp":10,
    "loc":"mum"
}], index=["emp5"])
#2.concatenate the new row with the existing dataframe
df=pd.concat([df,new_row])
print(df)
```

```python
#alternate method to add a new row (eaasier)
df.loc["emp6"]=["farq",50,10000,"agent",15,"jam"]
print(df)
```

```python
#adding multiple rows using the concat method
df1=pd.DataFrame({
    "name":["abdullah","husain","anas","moin"],
    "age":[15,20,12,10],
    "rollno":[29,11,2,3]
})
print(df1)
print("\n")

new_rows=pd.DataFrame([
    {
        "name":"hasnain",
        "age":12,
        "rollno":26
    },
    {
        "name":"husain",
        "age":13,
        "rollno":27
    }
    ], index=[4,5])
df1=pd.concat([df1,new_rows])
print(df1)

#to add multiple rows using the loc method ,just write the loc statement multiple times
```

```python
#practicing adding new column 

#adding a single new column
df1["marks"]=pd.DataFrame({
    "marks":[100,99,99,10,88,99]
})
print(df1)

#adding multiple new column
df1[["grade","result"]]=pd.DataFrame({
    "grade":["O","O","O","F","A","O"],
    "result":["pass","pass","pass","fail","pass","pass"]
})
print("\n")
print(df1)
```
