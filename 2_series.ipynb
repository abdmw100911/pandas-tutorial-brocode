```python
import pandas as pd
```

```python
data=pd.Series([1,2,3,4], index=["a","b","c","d"])
print(data)
```

```python
ls=[1,2,3,4,5]
ls1=pd.Series(ls,index=["a","b","c","d","e"])
print(ls1)
```

```python
serie=pd.Series([22,23,24,25,26],index=["a","b","c","d","e"])
print(serie)
#using the loc(access by label), to access and modify an element in the series
serie.loc["a"]=1000
print(serie.loc["a"])
print(serie)
```

```python
#using the iloc property to access and update the element in the serie 
serie.iloc[1]=69696
print(serie.iloc[1])
print(serie)
```
