```python
import pandas as pd
#series: a pandas "1D labelled array" that can hold "any data type" 
'''
0   100
1   200
2   300
^      ^
|      |
label  value
'''

data=[100,101,102,103] #list
print(type(data))

#converting the list into a series
series=pd.Series(data)
print(type(series))

print(series)
#this prints the data with labels and the metadata of the data (eg. int, float, bool)

data1=[100.1,100.2]
series1=pd.Series(data1)
print(series1)

data2=[True,False]
series2=pd.Series(data2)
print(series2)
```

```python
#the label of series is usually int, but it can be modified as well

series=pd.Series(data, index=["a","b","c","d"])
print(series)

series=pd.Series(data, index=["ap1","ap2","ap3","ap4"])
print(series)
```

```python
#accessing the elements of the series using the loc property
#1. loc->location by label
#2. iloc->location by int index

series=pd.Series(data, index=["a","b","c","d"])
print(series)

print(series.loc["a"])#prints 100

print(series.loc["d"])#prints 103

#using iloc
print(series.iloc[0])#prints 100

print(series.iloc[3])

#the values in the series can also be updated using the loc property
series.loc["a"]=200
print(series)

series.iloc[3]=800
print(series)
```

```python
#filtering by value 
age=pd.Series([10,12,69,18,25,37,55,60,65])
print("all the age groups that are not eligible to start work")
print(age[age<18])

print("all age groups that are eligible to work are")
print(age[(age >=18) & (age <60)])

print("all age groups that are retired are")
print(age[age>=60])
```

```python
#calorie example using dictionary for filtering
#give the days you stick to the diet ,(calorie<=2000) and the days you dont
#stick to the diet 

calorie={
    "day1":1000,
    "day2":5000,
    "day3":3000,
    "day4":1900

}
calorie=pd.Series(calorie)
print(calorie)

print("the days in which i sticked to my diet were")
print(calorie[calorie<=2000])

print("the days in which i didnt stick to my diet were")
print(calorie[calorie>2000])

print("mai sala bhais ki maut hi maruga")
```
